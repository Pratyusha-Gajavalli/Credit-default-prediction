```python
import pandas as pd
```

```python
Data = pd.read_csv('creditDefault_Train.csv')
Data.head()
```

# Data Preprocessing

```python
Data['Sex']=Data['Sex'].astype('str')
Data['Education']=Data['Education'].astype('str')
Data['Marriage']=Data['Marriage'].astype('str')
Data['Default']=Data['Default'].astype('str')
```

```python
#Counting missing values
missing_percents=dict()
for i in range(len(Data.columns)):
    missing_percents[Data.columns[i]]=round(Data[Data.columns[i]].isnull().sum() * 100 / len(Data),2)
    
missing_percents
```

```python
cols = list(Data.columns)
cols.remove('Default')
```

```python
#Creating dummies
Data_dummies = pd.get_dummies(data=Data, drop_first=True)
Data_dummies.head()
```

# Checking for Correlation

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
plt.figure(figsize=(20,10))
mask=np.triu(np.ones_like(Data_dummies.corr()))
heatmap=sns.heatmap(Data_dummies.corr(),vmin=-1,vmax=1,annot=True,cmap="BrBG")
heatmap.set_title("Correlation Heatmap")
```

```python
Data_dummies.corr()
```

# Feature Engineering

```python
Data_dummies_new = Data_dummies.copy()
Data_dummies_new['S1-P1'] = Data_dummies_new['Statement_1']-Data_dummies_new['Payment_1']
Data_dummies_new['S2-P2'] = Data_dummies_new['Statement_2']-Data_dummies_new['Payment_2']
Data_dummies_new['S3-P3'] = Data_dummies_new['Statement_3']-Data_dummies_new['Payment_3']
Data_dummies_new['S4-P4'] = Data_dummies_new['Statement_4']-Data_dummies_new['Payment_4']
Data_dummies_new['S5-P5'] = Data_dummies_new['Statement_5']-Data_dummies_new['Payment_5']
Data_dummies_new['S6-P6'] = Data_dummies_new['Statement_6']-Data_dummies_new['Payment_6']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_1']/Data_dummies_new['Limit']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_2']/Data_dummies_new['Limit']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_3']/Data_dummies_new['Limit']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_4']/Data_dummies_new['Limit']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_5']/Data_dummies_new['Limit']
Data_dummies_new['S1L'] = Data_dummies_new['Statement_6']/Data_dummies_new['Limit']
Data_dummies_new['Avg_Status'] = Data_dummies_new[['Status_1','Status_2','Status_3','Status_4','Status_5','Status_6']].mean(axis=1)
Data_dummies_new['Max_Status'] = Data_dummies_new[['Status_1','Status_2','Status_3','Status_4','Status_5','Status_6']].max(axis=1)
```

# Splitting into train and validation sets

```python
cols_dum = list(Data_dummies.columns)
cols_dum.remove('Default_1')
```

```python
Data_X = Data[cols]
Data_X_dum = Data_dummies[cols_dum]
Data_Y = Data['Default']

from sklearn.model_selection import train_test_split

# Stratified split of the train and test set with train-test ratio of 7:3
X_train, X_test, y_train, y_test = train_test_split(Data_X_dum, Data_Y, test_size=0.3, 
                                            stratify=Data_Y, random_state=10)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf_LR = LogisticRegression()

print('ROC AUC:', cross_val_score(clf_LR, Data_X, Data_Y, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
print('ROC AUC:', cross_val_score(gnb, X_train, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Support Vector 

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
df = X_train.copy()
df[cols] = StandardScaler().fit_transform(df[cols])
```

```python
from sklearn.svm import SVC
SV_classifier = SVC()
print('ROC AUC:', cross_val_score(SV_classifier, df, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=40)
print('ROC AUC:', cross_val_score(KNN, df, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Random Forest

```python
Data_X = Data[cols]
Data_Y = Data['Default']

from sklearn.model_selection import train_test_split

# Stratified split of the train and test set with train-test ratio of 7:3
X_train, X_test, y_train, y_test = train_test_split(Data_X, Data_Y, test_size=0.3, 
                                            stratify=Data_Y, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier

# See the inital model performance
RFC = RandomForestClassifier(random_state=10)

print('ROC AUC:', cross_val_score(RFC, X_train, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(n_estimators=65,max_depth=2,subsample=0.7,min_samples_leaf=0.0005,learning_rate=0.2,random_state=10)
print('ROC AUC:', cross_val_score(GBC, X_train, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
grid = {'n_estimators': list(range(40,70,5)),
        'learning_rate': [0.05, 0.1, 0.15, 0.2,0.25, 0.3],
        'max_depth': [2, 4, 6, 8]}

GBC2 = GradientBoostingClassifier(random_state=10)
CV_model = GridSearchCV(estimator=GBC2, param_grid=grid)
print('ROC AUC:', cross_val_score(CV_model, X_train, y_train, 
                                  cv=StratifiedKFold(n_splits=5), 
                                  scoring='roc_auc').mean())
```

# Prediction

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
GBC_final = GradientBoostingClassifier(n_estimators=65,max_depth=2,subsample=0.7,min_samples_leaf=0.0005,learning_rate=0.2, random_state=10)
GBC_final.fit(X_train, y_train)
predictions = GBC_final.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print()
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
predictions_prob = GBC_final.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, predictions_prob[:, 1])
```
